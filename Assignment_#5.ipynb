```python
import pandas as pd #import pandas
 #set df to the csv file
df = pd.read_csv('BDHSI2016_music_reviews.csv',encoding='mac_roman', sep='\t')

#preprocess the df
import nltk #imported nltk
lemmas = nltk.WordNetLemmatizer()   
#set lemmas to a function in nltk that allows lemmatization
body_tokens =  df['body'].apply(nltk.word_tokenize)  
#body_tokens stores tokenized 'body'
body_tokens_lems = []      
#initalize body_tokens_lems
for body in body_tokens:     
    #loops thru each element in body_token. This is the 'body' of each row
    body_tokens_lems.append([lemmas.lemmatize(t) for t in body]) 
    #appends the lemmatized/tokenized 'body' of each row


from nltk.corpus import stopwords   
#import stopwords
stopwords = stopwords.words('english') 
#the variable stopwords stores list of storewords

pre_processed_body = []
#initialize variable that eventually stores lemmatized/lowercased/non-stopwords version of body

for body in body_tokens_lems:     
    #loops thru each element in body_tokens_lems. This is the 'body' of each row
    words = [w.lower() for w in body if w.isalpha()] 
    #lowercase all the words/take out non-alphabetical characters
    words = [item for item in words if item not in stopwords] 
    #take away stopwords
    pre_processed_body.append(words)  
    #add the pre-processed body into a variable
df['pre_processed'] = pre_processed_body 
#make column of pre-processed body in df 


from nltk.tokenize.treebank import TreebankWordDetokenizer 
#import detokenizer function from nltk
df['pre_processed'] = df['pre_processed'].apply(TreebankWordDetokenizer().detokenize) 
#detokenize preprocessed body

df.head() #return df
```

First, I imported pandas library as pd, and set the csv file to df. Now, I need to preprocess the body of df before anything. So, I imported nltk library. From this, I tokenized and lemmatized the body. Next, I imported the stopwords from the nltk package. Then, I took out all the non-alphabetical characters, and lowercased all the alphabetical characters. Then, I removed the stopwords in the body. Then I created a column in df called 'pre-processed' that contains a tokenized, lemmatized, non-stopword body that is all lowercased and only contains alphabetical words of body. However, I want this column to be detokenized to processing this body later on. So I imported a detokenizer function from nltk and detokenized the 'pre_processed' column of df. 

Create a DTM(Document Term Matrix) weighted by the tf-idf score by using `TfidfVectorizer()`.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#import vectorizer to make the DTM
tfidfvec = TfidfVectorizer()
#set TfidfVectorizer() to tfidfvec

# To create a DTM, but with cells weigthed by the tf-idf score.
dtm_tfidf_df = pd.DataFrame(tfidfvec.fit_transform(df.pre_processed).toarray(), 
                            columns=tfidfvec.get_feature_names(), 
                            index = df.index) 
#created DTM using 'pre_processed' column of df


dtm_tfidf_df #returned the table
```

First, I imported the vectorizer functiuon from sklearn that will be used to make the DTM. Next, I created a DTM dataframe by using the vectorized 'pre-processed' column of df, the unique words of 'pre-processed' as columns, and used the same indexing as df. Lastly, I returned the dataframe.  


Top 20 words with the highest tf-idf weights:

```python
dtm_tfidf_df.max().sort_values(ascending=False)[:20]  
#Sorted dtm_tfidf_df in descending order and got the top 20 words of it
```

Using the DTM dataframe, I got the max values (td-idf weights) of each column. Then, I sorted this series by descending order so the largest td-idf weighted words will be the first element. Lastly, I printed to top 20 words of the series, which will be the 20 words with heighest weights.

Add a column named 'GENRE' by using the original 'genre' column from dataframe 'df' to the DTM

```python
df_type = df['genre'].to_frame() 
#created a dataframe of the 'genre' column of df
merged_df = df_type.join(dtm_tfidf_df, how = 'right', lsuffix='_x') 
#merged the 'genre' dataframe with DTM
merged_df #returned table
```

```python
grp_type = merged_df.groupby('genre_x') 
#grouped the merged dataframe by the genre
grp_type.max(numeric_only=True) 
#sorted columns by the max value in the grouped genre
```

I created a groupby dataframe by its respective genre of the DTM dataframe. Next, I got the max value of each column for the grouped genre.   

Top 10 words with the highest tf-idf weights for 'Indie', 'Folk', and 'Rap', respectively.

```python
print("Indie Words")
print(grp_type.get_group('Indie').max(numeric_only=True).sort_values(ascending=False)[0:10]) 
```

```python
print("\nFolk Words")
print(grp_type.get_group('Folk').max(numeric_only=True).sort_values(ascending=False)[0:10])
```

```python
print("\nRap Words")
print(grp_type.get_group('Rap').max(numeric_only=True).sort_values(ascending=False)[0:10])
```

Same process was repeated 3 times. First, I got the specified group/genre, so for the first one, it will be 'Indie'. Next, I got the max tf-idf values of each column of that group/genre, and sorted the retrieved series in descending order. Lastly, I printed the first 10 items of the series, whereby this represents the top 10 words with the highest weight in 'Indie' genre. This was repeated for 'Folk' and 'Rap.'
